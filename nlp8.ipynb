```python
def nlp80():
    with codecs.open("enwiki-20150112-400-r100-10576.txt", "r", encoding="utf-8") as line:
        text = [i.strip().split() for i in line.readlines() if len(i.strip()) > 0]
    for sentence in tqdm(text):
        sen_len = len(sentence)
        pre_sen = sentence
        for wd in sentence:
            pre_wd = wd
            wd = re.sub(r"^[\.\,\!\?\;\:\(\)\[\]\'\"]+(\w.+\w)[\.\,\!\?\;\:\(\)\[\]\'\"]+$",r"\1",wd)
        sentence = [wd for wd in sentence if len(wd)> 0]
    text = "\n".join([" ".join(sentence) for sentence in text])
    with codecs.open("nlp8.txt", "w", encoding="utf-8") as f:
        f.write(text)
    print("finished")
    
if __name__ == "__main__":
    nlp80()
```

```python
def nlp81():
    pattern = re.compile(r"(.*)(Bosnia and Herzegovina|Burkina Faso|Central African Republic|Costa Rica|"+ \
    "Czech Republic|Dominican Republic|East Timor|El Salvador|Equatorial Guinea"+  \
    "|New Zealand|Papua New Guineaa|St Kitts and Nevis|St Vincent and the Grenadines|San Marino" + \
  "|São Tomé and Príncipe|Saudi Arabia|Sierra Leone|Solomon Islands|South Africa"+ \
  "|South Sudan|Sri Lanka|Trinidad and Tobago|United Arab Emirates|United Kingdom|United States)(.*)")
    with codecs.open("nlp8.txt", "r", encoding="utf-8") as f:
        text = [i.strip() for i in f.readlines() if len(i.strip()) > 0]
    for i, line in enumerate(text):
        if pattern.match(line):
            text[i] = pattern.match(line)[1] + \
                    re.sub(" ","_", pattern.match(line)[2]) + \
                    pattern.match(line)[3]
    text = "\n".join(["".join(line) for line in text])
    with codecs.open("nlp8_name_changed.txt", "w", encoding="utf-8") as f:
        f.write(text)
    print("finished")
if __name__ == "__main__":
    print(nlp81())
```

```python
def nlp82(vervose = False):
    with codecs.open("nlp82_result.txt", "w", encoding="utf-8") as result_f:
        with codecs.open("nlp8_name_changed.txt", "r", encoding="utf-8") as f:
            for i in tqdm(f):
                wd_list = i.split()
                for ind, wd in enumerate(wd_list):
                    if wd == "United_States":
                        print(wd)
                    width = np.random.randint(1,5) 
                    if vervose:
                        print("width:", width)
                    result_f.write( wd + "\t" + "\t".join(wd_list[max(0, ind - width) : ind] + \
                                                wd_list[ind + 1: min(len(wd_list), ind + width + 1)]) + "\n")
    
                
if __name__ == "__main__":
    nlp82()
    with codecs.open("nlp82_result.txt", "r", encoding="utf-8") as result_f:
        for i, j in enumerate(result_f):
            if i < 10:
                print(j)
            else:
                break
```

```python
def nlp83(t, c):
    _sum_oc = 0
    flg = 0
    ct_t_c = Counter()
    ct_c = Counter()
    ct_t = Counter()
    with codecs.open("nlp82_result.txt", "r", encoding="utf-8") as result_f:
        for i in tqdm(result_f):
            _t = i.split()[0]
            _cs = i.split()[1:]
            ct_t[_t] += 1
            for _c in _cs:
                ct_c[_c] += 1
                ct_t_c[(_t, _c)] += 1
            _sum_oc += 1
    pickle.dump( ct_t_c, open("ct_t_c", "wb"))
    pickle.dump( ct_t, open("ct_t", "wb"))
    pickle.dump( ct_c, open("ct_c", "wb"))
    return ct_t_c[(t,c)], ct_t[t], ct_c[c], _sum_oc 

if __name__ == "__main__":
    print(nlp83("political", "is"))
```

```python
def nlp84():
    wd_ls = []
    ct_t_c = pickle.load(open("ct_t_c","rb"))
    ct_t = pickle.load(open("ct_t","rb"))
    ct_c = pickle.load(open("ct_c","rb"))
    t_ls = list(ct_t.keys())
    c_ls = list(ct_c.keys())
    PRMI = sparse.lil_matrix((len(t_ls), len(c_ls)))
    _sum_oc = len(ct_t_c)
    for (_t, _c), count in tqdm(ct_t_c.items()):
        if count > 10:
            cor_t_c = count
            oc_t = ct_t[_t]
            oc_c = ct_c[_c]
            ind_t = t_ls.index(_t)
            ind_c = c_ls.index(_c)
            PRMI[ind_t, ind_c] = max(np.log(_sum_oc * cor_t_c / oc_t / oc_c), 0)
    pickle.dump(PRMI, open("PRMI", "wb"))
    return PRMI


if __name__ == "__main__":
    print(nlp84()[:100, :100])
```

```python
def nlp87():
    PRMI = pickle.load(open("decomposed_PRMI", "rb"))
    ct_t = pickle.load(open("ct_t","rb"))
    t_ls =list(ct_t.keys())
    u_s_ind = t_ls.index("U.S.")
    U_S_ind = t_ls.index("United_States")
    dot = PRMI[u_s_ind, :].dot(PRMI[U_S_ind, :].T)
    return dot / np.linalg.norm(PRMI[u_s_ind]) / np.linalg.norm(PRMI[U_S_ind]) 

if __name__ == "__main__":
    print(nlp87())
```

```python
def nlp88():
    PRMI = pickle.load(open("decomposed_PRMI", "rb"))
    ct_t = pickle.load(open("ct_t","rb"))
    t_ls =list(ct_t.keys())
    eng_ind =  t_ls.index("England")
    eng_vec = PRMI[eng_ind,:] 
    eng_vec = eng_vec / np.linalg.norm(eng_vec)
    PRMI_norm = np.linalg.norm(PRMI.T, axis = 0)
    result = eng_vec.dot(PRMI.T / PRMI_norm)
    t_ls = np.array(t_ls)[~np.isnan(result)]
    result = result[~np.isnan(result)]
    top_10 = np.argsort(result)[-10:][::-1]
    top_dict = {t_ls[i] : result[i] for i in top_10}
    return top_dict

if __name__ == "__main__":
    print(nlp88())
```

```python
def nlp89():
    PRMI = pickle.load(open("decomposed_PRMI", "rb"))
    ct_t = pickle.load(open("ct_t","rb"))
    t_ls =list(ct_t.keys())
    spa_ind =  t_ls.index("Spain")
    mad_ind =  t_ls.index("Madrid")
    ath_ind =  t_ls.index("Athens")
    target_vec = PRMI[spa_ind, :] - PRMI[mad_ind, :] + PRMI[ath_ind, :]
    target_vec = target_vec / np.linalg.norm(target_vec)
    PRMI_norm = np.linalg.norm(PRMI.T, axis = 0)
    result = target_vec.dot(PRMI.T / PRMI_norm)
    t_ls = np.array(t_ls)[~np.isnan(result)]
    result = result[~np.isnan(result)]
    top_10 = np.argsort(result)[-10:][::-1]
    top_dict = {t_ls[i] : result[i] for i in top_10}
    return top_dict

if __name__ == "__main__":
    print(nlp89())
```
