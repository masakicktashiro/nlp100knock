```python
import re
import codecs
from nltk.stem import PorterStemmer
import xml.etree.ElementTree as ET
import pydot_ng as pydot
```

```python
def nlp50():
    with codecs.open("train/text/nlp.txt", "r", encoding="utf-8") as f:
        text = f.read()
    text = re.sub("\n", " ", text)
    while True:
        match = re.match(r".*?([A-Z].*?)(\.|;|:|\?|\!)\s+?[A-Z]", text, re.MULTILINE + re.DOTALL)
        if match is None:
            break
        text  = text[match.end(2):]
        yield match.group(1) + match.group(2)
        
        

if __name__ == "__main__":
    for i, j in enumerate(nlp50()):
        print(i, j)
```

```python
def nlp51():
    sentences = nlp50()
    for sentence in sentences:
        for wd in sentence.split():
            wd = re.sub(r"\W","",wd)
            yield wd
        yield "\n"
        
        
if __name__ == "__main__":
    for i, j in enumerate(nlp51()):
        if i < 100:
            print(j)
```

```python
def nlp52():
    stemmer = PorterStemmer()
    words = nlp51()
    for wd in words:
        yield wd + "\t" + stemmer.stem(wd)
        
        
if __name__ == "__main__":
    for i, j in enumerate(nlp52()):
        if i < 100:
            print(j)
```

```python
def nlp53():
    file_path = "train/text/nlp.txt.xml"
    tree = ET.parse(file_path)
    root = tree.getroot()
    for i in root.iter("word"):
        yield i.text
        
        
if __name__ == "__main__":
    for i, j in enumerate(nlp53()):
        if i < 30:
            print(j)
```

```python
def nlp54():
    file_path = "train/text/nlp.txt.xml"
    tree = ET.parse(file_path)
    root = tree.getroot()
    for i in root.getiterator():
        if i.tag == "token":
            result = []
        if i.tag in ["word","lemma"]:
            result.append(i.text)
        if i.tag == "POS":
            result.append(i.text)
            yield "\t".join(result)
    

if __name__ == "__main__":
    for i in nlp54():
        print(i)
```

```python
def nlp56():
    text = []
    file_path = "train/text/nlp.txt.xml"
    tree = ET.parse(file_path)
    root = tree.getroot()
    replace_dict = dict()
    replace_ls = []
    sub_ls = []
    for i in root.iter("coreference"):
        if i[0].tag == "mention":
            replace_dict[int(i[0].find("start").text)] = \
                              i[0].find("text").text + "(" +i[1].find("text").text + ")"
            
            replace_ls.append([int(i[0].find("sentence").text),
                              int(i[0].find("start").text)])
            for ind in range(int(i[0].find("start").text) + 1, 
                            int(i[0].find("end").text) + 1):
                sub_ls.append([int(i[0].find("sentence").text), ind])
    for i in root.iter("sentence"):
        if len(i.attrib) == 0:
            continue
        sentence_id = int(i.attrib["id"])
        for j in i.find("tokens").findall("token"):
            wd_id = int(j.attrib["id"])
            if [sentence_id, wd_id] in sub_ls:
                continue
            elif [sentence_id, wd_id] not in replace_ls:
                text.append(j.find("word").text)
                continue
            else:
                text.append(replace_dict[wd_id])
        text.append("\n")
    text = " ".join(text)
    text = re.sub(r"(\w)\s(\.|;|:|\?|\!|,)", r"\1" + r"\2", text)
    return text


if __name__ == "__main__":
    print(nlp56())
```

```python
def nlp57():
    file_path = "train/text/nlp.txt.xml"
    tree = ET.parse(file_path)
    root = tree.getroot()
    for h, i in enumerate(root.iter("sentence")):
        graph = pydot.Dot(graph_type='digraph')
        graph.add_node(pydot.Node(0, label="root"))
        if len(i.attrib) == 0:
            continue
        token_ls = []

        for j in i.find("tokens").findall("token"):
            if j.find("word").text == ",":
                continue
            graph.add_node(pydot.Node(int(j.attrib["id"]),
                                      label=j.find("word").text))

        for ind, dep in enumerate(i.find("dependencies[@type='collapsed-dependencies']"), 1):
            graph.add_edge(pydot.Edge(int(dep.find("governor").attrib["idx"]), 
                                     int(dep.find("dependent").attrib["idx"])))
        graph.write_png("train/pic/eng_result%d.png"%h)
        

if __name__ == "__main__":
    nlp57()
```

```python
def nlp58():
    file_path = "train/text/nlp.txt.xml"
    tree = ET.parse(file_path)
    root = tree.getroot()
    for h, i in enumerate(root.iter("sentence")):
        nsubj_parent_child_dict = dict()
        dobj_parent_child_dict = dict()
        if len(i.attrib) == 0:
            continue
        for ind, dep in enumerate(i.find("dependencies[@type='collapsed-dependencies']"), 1):
            if dep.attrib["type"] == "nsubj":
                nsubj_parent_child_dict[dep.find("governor").text] = \
                                             dep.find("dependent").text
            if dep.attrib["type"] == "dobj":
                dobj_parent_child_dict[dep.find("governor").text] = \
                                             dep.find("dependent").text
        nsubj_parent_set = {i for i in nsubj_parent_child_dict.keys()}
        dobj_parent_set = {i for i in dobj_parent_child_dict.keys()}
        if len(list(nsubj_parent_set.intersection(dobj_parent_set))) == 0:
            continue
        verb = list(nsubj_parent_set.intersection(dobj_parent_set))[0]
        subject = nsubj_parent_child_dict[verb]
        _object = dobj_parent_child_dict[verb]
        yield subject + "\t" + verb + "\t" + _object
        
        
if __name__ == "__main__":
    for i in nlp58():
        print(i)
```

```python
def nlp59():
    file_path = "../stanford-corenlp-full-2018-02-27/nlp.txt.xml"
    tree = ET.parse(file_path)
    root = tree.getroot()
    for i in root.iter("parse"):
        s_shiki = i.text
        while re.match(r".*?(\(NP.+)", s_shiki) is not None:
            s_shiki = re.match(r".*?(\(NP.+)", s_shiki).group(1)
            match = regex.match(r"(?<rec>\((?:[^()]+|(?&rec))*\))", s_shiki)
            NP = match.group(1)
            temp = NP
            while re.match(r".+?(\(NP.+)", temp) is not None:
                temp = re.match(r".+?(\(NP.+)", temp).group(1)
                sub_NP = regex.match(r"(?<rec>\((?:[^()]+|(?&rec))*\))", temp).group(1)
                sub_NP = re.sub(r"(\(\S+|\))", "", sub_NP)
                sub_NP = re.sub(r"\s+"," ", sub_NP)
                sub_NP = re.sub(r"(\w)\s(\.|;|:|\?|\!|,)", r"\1" + r"\2", sub_NP)
                yield sub_NP
            s_shiki = s_shiki[len(NP):]
            NP = re.sub(r"(\(\S+|\))", "", NP)
            NP = re.sub(r"\s+"," ", NP)
            NP = re.sub(r"(\w)\s(\.|;|:|\?|\!|,)", r"\1" + r"\2", NP)
            yield NP
        

if __name__ == "__main__":
    for i in nlp59():
        print(i)
```
